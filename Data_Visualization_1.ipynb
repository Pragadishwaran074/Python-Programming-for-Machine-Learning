```python
import matplotlib.pyplot as plt
import numpy as np
xpoints = np.array([0,5])
ypoints = np.array([10,100])
print(plt.plot(xpoints,ypoints))
print(plt.show())
```

```python
x = np.array([0,3])
y = np.array([0,9])
print(plt.plot(x,y,marker='p'))
print(plt.show())
```
